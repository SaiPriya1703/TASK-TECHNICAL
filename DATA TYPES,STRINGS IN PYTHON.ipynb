DATA TYPES
1.List

```python
stud=['nikitha','sai priya','kiran']
print(stud)
print(type(stud))
stud.append('shila')
print(stud)
stud.insert(1,'naresh')
print(stud)
stud.remove('naresh')
print(stud)
stud.pop(0)
print(stud)
stud.sort()
print(stud)
stud.reverse()
print(stud)
l=len(stud)
print(l)
fruit='Banana'
fruit.count('a')
```

2.Tuple

```python
car_brands=('Benz','suzuki','audi','audi')
print(car_brands)
print(len(car_brands))
print(car_brands.count('audi'))
print(car_brands[1])
```

3.Set

```python
name1={'sai','priya','nikki'}
name2={'nikki','saki','teju'}
name1.add('sakshi')
print(name1)
name1.remove('sakshi')
print(name1)
unions=name1.union(name2)
print(unions)
inter=name1.intersection(name2)
print(inter)
diff=name1.difference(name2)
print(diff)
name1.clear()
print(name1)
```

4.Dictionary

```python
student = {
    "name": "Sai Priya",
    "age": 21,
    "course": "B.Tech"
}
print("Keys:", student.keys())
print("Values:", student.values())
print("Items:", student.items())
print("Name:", student.get("name"))
student.update({"age": 22})
print("After update:", student)
student.pop("course")
print("After pop:", student)
```

STRINGS IN PYTHON

```python
s='python programming'
print(len(s))
print(s.upper())
print(s.lower())
print(s.capitalize())
print(s.title())
print(s.replace('python','java'))
print(s.split())
print(s.find('programming'))
print(s.count('o'))
print(s.startswith('python'))
print(s.endswith('programming'))
s1='python'
print(s1)
print(len(s1))
print(s1.strip())

```
