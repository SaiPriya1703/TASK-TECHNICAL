```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="year", y="month", data=flights,hue="month")
plt.title("Scatter Plot-Year vs Passengers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="month", y="passengers", data=flights)
plt.title("Average Passengers by Month")
plt.show()
```
