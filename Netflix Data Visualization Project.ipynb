```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

1.Which type of content is more available on Netflix?

```python
plt.figure(figsize=(6,5))

sns.countplot(x='type', data=df)

plt.title("Movies vs TV Shows")
plt.xlabel("Content Type")
plt.ylabel("Count")

plt.show()
```

2.Which countries contribute the most content to Netflix?

```python
top_country = df['country'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_country.values, y=top_country.index)

plt.title("Top Countries Contributing Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")

plt.show()
```

3.Which directors have the most titles on Netflix?

```python
top_directors = df['director'].value_counts().head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    x=top_directors.values,
    y=top_directors.index
)

plt.title("Top 10 Directors on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Director")
plt.show()
```
